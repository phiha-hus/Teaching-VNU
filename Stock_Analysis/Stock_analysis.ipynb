```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2021,8,18)
```

```python
google = web.DataReader("GOOGL",'yahoo',start,end)
google
#vos = web.DataReader("VOS",'ssi',start,end)
#vos
```

```python
def inc_dec(c, o):
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value

google["Status"]=[inc_dec(c,o) for c, o in zip(google.Close,google.Open)]
google
```
