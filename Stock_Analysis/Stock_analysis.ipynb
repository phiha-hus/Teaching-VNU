```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2021,9,1)
end = datetime.datetime(2021,9,30)
```

```python
google = web.DataReader("GOOGL",'yahoo',start,end)
google
#vos = web.DataReader("VOS",'ssi',start,end)
#vos
```

```python
google.head()
```

```python
google.describe()
```

```python
def inc_dec(c, o):
    if c > o:
        value="Increase"
    elif c < o:
        value="Decrease"
    else:
        value="Equal"
    return value

google["Status"]=[inc_dec(c,o) for c, o in zip(google.Close,google.Open)]
google["Middle"]=(google.Open + google.Close)/2
google["Height"]=abs(-google.Open + google.Close)
google
```

```python
from bokeh.plotting import figure, show, output_file

p = figure(x_axis_type='datetime', width=1000, height=300)
p.title.text = "Candlestick Chart"

hours_12 = 12 * 60 * 60 * 1000

#tăng
p.rect(google.index[google.Status=="Increase"],google.Middle[google.Status=="Increase"],
   hours_12, google.Height[google.Status=="Increase"],fill_color="#CCFFFF",
   line_color="black")
#giảm
p.rect(google.index[google.Status=="Decrease"],google.Middle[google.Status=="Decrease"],
   hours_12, google.Height[google.Status=="Decrease"],fill_color="#FF3333",
   line_color="black")

p.segment(google.index, google.High, google.index, google.Low, color="Black")

output_file("chart.html")
show(p)
```
