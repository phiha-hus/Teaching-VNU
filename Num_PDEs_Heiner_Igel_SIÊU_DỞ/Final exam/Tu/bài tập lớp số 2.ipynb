## Câu 1: 
Xét phương trình phương trình sóng elastic 1D có dạng:
\begin{equation}
\rho(x) \partial_t^2 u(x,t) = \partial_x (\mu(x) \partial_x u(x,t)) + f(x,t).
\end{equation}

Nhân cả hai vế của phương trình trên với $\varphi_j$ rồi lấy tích phân hai vế trên toàn miền, ta được dạng yếu của phương trình sóng:

\begin{equation}
\int_D \rho(x) \partial_t^2 u(x,t) \varphi_j \, dx = \int_D \partial_x (\mu(x) \partial_x u(x,t)) \varphi_j \, dx + \int_D f(x, t) \varphi_j \, dx.
\end{equation}

Sử dụng công thức tích phân từng phần, ta có
\begin{equation}
\int_D \partial_x (\mu(x) \partial_x u(x,t)) \varphi_j \, dx = \left( \mu \partial_x u (x, t) \varphi_j \right) - \int_D \mu \partial_x u(x, t) \partial_x \varphi_j \, dx.
\end{equation}
Áp dụng điều kiện biên tự do nên 
\begin{equation}
\left( \mu \partial_x u (x, t) \varphi_j \right)|_D=0
\end{equation}
Do đó
\begin{equation}
\int_D \rho(x) \partial_t^2 u(x,t) \varphi_j \, dx + \int_D \mu \partial_x u(x, t) \partial_x \varphi_j \, dx = \int_D f(x, t) \varphi_j \, dx.
\end{equation}
Với hàm 
$$ u(x, t) \rightarrow \overline{u}(x, t) = \sum\limits_{i = 1}^N u_i(t) \varphi_i(x) $$
Khi đó, phương trình trên trở thành
\begin{eqnarray}
&& \int_D \rho(x) \partial_t^2 \overline{u}(x,t) \varphi_j \, dx + \int_D \mu \partial_x \overline{u}(x, t) \partial_x \varphi_j \, dx = \int_D f(x, t) \varphi_j \, dx\\
& \Leftrightarrow & \int_D \rho(x) \partial_t^2 \sum\limits_{i = 1}^N u_i(t) \varphi_i(x)  \varphi_j \, dx + \int_D \mu \partial_x \sum\limits_{i = 1}^N u_i(t) \varphi_i(x)  \partial_x \varphi_j \, dx = \int_D f(x, t) \varphi_j \, dx \\
& \Leftrightarrow & \sum\limits_{i = 1}^N \partial_t^2 u_i(t) \int_D \rho(x)  \varphi_i(x)  \varphi_j \, dx + \sum\limits_{i = 1}^N u_i(t) \int_D \mu \partial_x \varphi_i(x)  \partial_x \varphi_j \, dx = \int_D f(x, t) \varphi_j \, dx.
\end{eqnarray}
Với

\begin{eqnarray}
&& u(t) \rightarrow u_i(t)\\
&& M \rightarrow M_{ij} = \int_D \rho \varphi_i \varphi_j \, dx \\
&& K \rightarrow K_{ij} = \int_D \partial_x \varphi_i \partial_x \varphi_j \, dx \\
&& f \rightarrow f_{j} = \int_D f(x, t) \varphi_j \, dx 
\end{eqnarray}
Suy ra, ta thu được phương trình dạng
$$ M^T \partial_t^2 u + K^T u = f.$$

Sử dụng công thức sai phân

$$ \partial_t^2 u \approx \frac{u(t + dt) - 2u(t) + u(t - dt)}{dt^2}$$
ta suy ra
$$ M^T \left[ \frac{u(t + dt) - 2u(t) + u(t - dt)}{dt^2} \right] = f - K^T u $$
hay
\begin{equation}
\mathbf{u}(t + dt) = dt^2 (\mathbf{M}^T)^{-1}[\mathbf{f} - \mathbf{K}^T\mathbf{u}] + 2\mathbf{u} - \mathbf{u}(t-dt).
\end{equation}

Gọi $\xi = x - x_i$, $h = x_{i + 1} - x_i$ và định nghĩa
$$ \varphi_i(\xi) = \left\lbrace \begin{array}{ll}
\frac{\xi}{h_{i - 1}}+ 1 & -h_{i - 1} \leq \xi \leq 0 \\ 
1 - \frac{\xi}{h_i} & 0 \leq \xi \leq h_i \\
0 & nếu  khác.
\end{array}  \right. $$

Do đó

$$ \partial_\xi \varphi_i(\xi) = \left\lbrace \begin{array}{ll}
\frac{1}{h_{i - 1}} & -h_{i - 1} \leq \xi \leq 0 \\ 
- \frac{1}{h_i} & 0 \leq \xi \leq h_i \\
0 & nếu  khác.
\end{array}  \right. $$

Thực hiện phép đổi biến số theo $\xi$, ta thu được

$$ M_{ii}  = \int_{D_\xi} \rho \varphi_i \varphi_i \, d\xi = \rho\int_{-h_{i - 1}}^0 \left( \frac{\xi}{h_{i - 1}} + 1\right)^2 d\xi + \rho\int_{0}^{h_i} \left( 1 - \frac{\xi}{h_{i}}\right)^2 d\xi = \frac{1}{3} \left( \rho h_{i - 1} + \rho h_i \right).$$

$$ M_{i, i-1} = \rho \int_{-h_{i - 1}}^0 \left( \frac{\xi}{h_{i - 1}} + 1 \right) \frac{-\xi}{h_{i - 1}}\, d\xi = \frac{1}{6} \rho h_{i - 1} $$
và 
$$ M_{i, i+1} = \rho \int_{0}^{h_i} \left(1 - \frac{\xi}{h_{i}} \right) \frac{\xi}{h_{i}}\, d\xi = \frac{1}{6} \rho h_{i} $$

Suy ra, trong trường hợp bước đi $h$ đều thì mass matrix $M$ có dạng

\begin{equation}
M = M_{ij} = \frac{\rho h}{6}
 \begin{pmatrix}
   \ddots  &    &    &    & 0\\
   1 & 4 &  1 &    &  \\
     & 1 &  4 &  1 &  \\
     &   &  1 &  4 & 1\\
   0 &   &    &    &  \ddots
 \end{pmatrix} 
\end{equation}

Tuy nhiên, trong câu hỏi này, bước đi $h$ không đều với

$$ h = h_i = \left\lbrace \begin{array}{ll}
40 & 1 \leq i \leq 115 \\ 
10 & 116 \leq i \leq 216 \\
20 & 217 \leq i \leq 447.
\end{array}  \right. $$

Do đó, ma trận $M$ có dạng

\begin{equation}
M = M_{ij} = \frac{\rho}{6}
 \begin{pmatrix}
   \ddots  &    &    &    & 0\\
   h_{i - 1} & 2h_{i - 1} + 2h_i &  h_i &    &  \\
     & h_{i} & 2h_{i} + 2h_{i + 1} &  h_{i + 1} &  \\
     &   &  h_{i + 1} & 2h_{i + 1} + 2h_{i + 2} &  h_{i + 2}\\
   0 &   &    &    &  \ddots
 \end{pmatrix} 
\end{equation}



### Stiffness matrix

Dạng tổng quát của ma trận độ cứng là

\begin{equation}
K = K_{ij} = \int_{D} \mu \partial_x\varphi_i \partial_x\varphi_j dx = \int_{D_{\xi}} \mu \partial_\xi\varphi_i \partial_\xi\varphi_j d\xi
\end{equation} 

Suy ra

$$ K_{ii}  = \int_{D_{\xi}} \mu \partial_\xi\varphi_i \partial_\xi\varphi_i d\xi = \mu \int_{-h_{i - 1}}^0 \left( \frac{1}{h_{i - 1}}\right)^2 d\xi + \mu \int_{0}^{h_i} \left( \frac{1}{h_{i}}\right)^2 d\xi = \frac{\mu}{h_{i - 1}} + \frac{\mu}{h_{i}}, $$
$$ K_{i, i - 1} = \int_{D_{\xi}} \mu \partial_\xi\varphi_i \partial_\xi\varphi_{i - 1} d\xi = -\frac{\mu}{h_{i - 1}},$$
$$ K_{i, i + 1} = \int_{D_{\xi}} \mu \partial_\xi\varphi_i \partial_\xi\varphi_{i + 1} d\xi = -\frac{\mu}{h_{i}},$$

Do đó, ma trận độ cứng cần tìm có dạng

\begin{equation}
K = K_{ij} = \mu
 \begin{pmatrix}
   \ddots  &    &    &    & 0\\
  -h_{i - 1} & h_{i - 1} + h_i & -h_i &    &  \\
     & -h_{i} & h_i + h_{i + 1} & -h_{i + 1} &  \\
     &   & -h_{i + 1} & h_{i + 1} + h_{i + 2} & -h_{i + 2} \\
   0 &   &    &    &  \ddots
 \end{pmatrix} 
\end{equation}



```python
# Import all necessary libraries, this is a configuration step for the exercise.
# Please run it before the simulation code!
import numpy as np
import matplotlib
# Show Plot in The Notebook
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

### 1. Initialization of setup

```python

# Initialization of setup
# ---------------------------------------------------------------
# Basic parameters
nx1 = 116
nx2 = 100 
nx3 = 230
nx    = nx1+nx2+nx3    # Number of collocation points  
xmax  = 10200.  # Physical dimension [m]
vs1    = 6000   # Wave velocity [m/s] 
vs2 = 1500
vs3 = 3000
vs= np.arange(0,nx)
for i in range(0,nx1):
    vs[i] = vs1
for i in range(nx1,nx1+nx2):
    vs[i] = vs2
for i in range(nx1+nx2,nx):
    vs[i] = vs3
## vs = 3000
ro0   = 2500    # Density [kg/m^3]
nt    = 18000   # Number of time steps
isx   = 223     # Source location [m] 
eps   = 0.5     # Stability limit
iplot = 20     # Snapshot frequency

#  dx = xmax/(nx-1)          
dx1 = 40     # calculate space increment
dx2 = 10
dx3 = 20
dx= np.arange(0,nx)
for i in range(0,nx1):
    dx[i] = dx1
for i in range(nx1,nx1+nx2):
    dx[i] = dx2
for i in range(nx1+nx2,nx):
    dx[i] = dx3
    
# initialize space coordinates
# x  = np.arange(0, nx)*dx
x  = np.arange(0, nx)
for i in range(0, nx1):
    x[i] = x[i]*dx[i]
for i in range(nx1, nx1+nx2):
    x[i] = x[nx1-1] +(i-nx1+1)*dx[i] 
for i in range(nx1+nx2, nx):
    x[i] = x[nx1+nx2-1]+ (i-nx1-nx2+1)*dx[i]
x  = np.transpose(x)

h = np.diff(x)  # Element sizes [m]

# parameters   
ro = x*0 + ro0

mu =  ro*vs**2
# time step from stabiity criterion
dt = 0.5*eps*np.min(dx[i])/np.max(np.sqrt(mu[i]/ro[i]))

# initialize time axis
t   = np.arange(1, nt+1)*dt 

# ---------------------------------------------------------------
# Initialize fields
# ---------------------------------------------------------------
u    = np.zeros(nx)
uold = np.zeros(nx)
unew = np.zeros(nx)

p    = np.zeros(nx)
pold = np.zeros(nx)
pnew = np.zeros(nx)
```

### 2. Source time function

In 1D the propagating signal is an integral of the source time function. As we look for a Gaussian waveform, we initialize the source time function $f(t)$ using the first derivative of a Gaussian function.

\begin{equation}
f(t) = -\dfrac{2}{\sigma^2}(t - t_0)e^{-\dfrac{(t - t_0)^2}{\sigma^2}}
\end{equation}

Initialize a source time function called 'src'.  Use $\sigma = 20 dt$ as Gaussian width, and time shift $t_0 = 3\sigma$. Then, visualize the source in a given plot.

```python
# Initialization of the source time function
# ---------------------------------------------------------------
pt  = 20*dt     # Gaussian width
t0  = 3*pt      # Time shift
src = -2/pt**2 * (t-t0) * np.exp(-1/pt**2 * (t-t0)**2)

# Source vector
# ---------------------------------------------------------------
f = np.zeros(nx); f[isx:isx+1] = f[isx:isx+1] + 1.

# ---------------------------------------------------------------
# Plot source time fuction
# ---------------------------------------------------------------
plt.plot(t, src, color='b', lw=2, label='Source time function')
plt.ylabel('Amplitude', size=16)
plt.xlabel('time', size=16)
plt.legend()
plt.grid(True)
plt.show()
```

### 3. The Mass Matrix
Having implemented the desired source, now we initialize the mass and stiffness matrices. In general, the mass matrix is given by

\begin{equation}
M_{ij} = \int_{D} \rho \varphi_i \varphi_j dx = \int_{D_{\xi}} \rho \varphi_i \varphi_j d\xi
\end{equation}



```python
# Mass matrix M_ij
# ---------------------------------------------------------------
M = np.zeros((nx,nx), dtype=float)
for i in range(1, nx-1):
    for j in range (1, nx-1):
        if j==i:
            M[i,j] = (ro[i-1]*h[i-1] + ro[i]*h[i])/3
        elif j==i+1:
            M[i,j] = ro[i]*h[i]/6
        elif j==i-1:
            M[i,j] = ro[i-1]*h[i-1]/6
        else:
            M[i,j] = 0
            
# Corner elements
M[0,0] = ro[0]*h[0]/3
M[nx-1,nx-1] = ro[nx-1]*h[nx-2]/3
# Invert M
Minv = np.linalg.inv(M)
```

### 4. The Stiffness matrix
The general form of the stiffness matrix is

\begin{equation}
K_{ij} = \int_{D} \mu \partial_x\varphi_i \partial_x\varphi_j dx = \int_{D_{\xi}} \mu \partial_\xi\varphi_i \partial_\xi\varphi_j d\xi
\end{equation} 



```python
# Stiffness matrix Kij
# ---------------------------------------------------------------
K = np.zeros((nx,nx), dtype=float)
for i in range(1, nx-1):
    for j in range(1, nx-1):
        if i==j:
            K[i,j] = mu[i-1]/h[i-1] + mu[i]/h[i]
        elif i==j+1:
            K[i,j] = -mu[i-1]/h[i-1]
        elif i+1==j:
            K[i,j] = -mu[i]/h[i]
        else:
            K[i,j] = 0

K[0,0] = mu[0]/h[0]
K[nx-1,nx-1] = mu[nx-1]/h[nx-2]
```

```python
# Display finite element matrices
# ---------------------------------------------------------------
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(K[1:10,1:10])
ax1.set_title('Finite-Element Stiffness Matrix $\mathbf{K}$')
ax1.axis("off")

ax2.imshow(Minv[1:10,1:10])
ax2.set_title('Finite-Element Mass Matrix $\mathbf{M^{-1}}$')
ax2.axis("off")

plt.tight_layout()
plt.show()
```

### 6. Finite element solution 

Finally we implement the finite element solution using the computed mass $M$ and stiffness $K$ matrices together with a finite differences extrapolation scheme

\begin{equation}
\mathbf{u}(t + dt) = dt^2 (\mathbf{M}^T)^{-1}[\mathbf{f} - \mathbf{K}^T\mathbf{u}] + 2\mathbf{u} - \mathbf{u}(t-dt).
\end{equation}

# Câu 2
Ta biết rằng $$ u(x, t) = \sum\limits_{i = 1}^N u_i(t) \varphi_i(x) = \sum\limits_{i=1}^N c_i \varphi_i $$
Ta sẽ thay thế các hàm cơ bản $\varphi_i$ bởi các hàm định dạng bậc hai.

Bằng cách đặt $ \xi = \dfrac{x-x_i}{x_{i+1}-x_i}$ thì lúc này $\xi \in \left[0,1\right] $

Giả sử rằng độ lệch được mô tả dưới dạng một hàm bậc hai $u(\xi) = c_1 + c_2 \xi + c_3 \xi^2$, đồng thời chúng ta thêm một node nữa vào giữa phần tử, tức là  $\xi_{1, 2, 3} = 0, 1/2, 1$.

Cùng với những điểm node này ta sẽ thu được:
$$ \left\lbrace \begin{array}{l}
u_1 = c_1 \\ 
u_2 = c_1 + 0.5 c_2 + 0.25 c_3 \\
u_3 = c_1 + c_2 + c_3.
\end{array}  \right. $$
Viết dưới dạng ma trận ta được:
$$ u = \left[ \begin{array}{c}
u_1 \\ 
u_2 \\
u_3
\end{array}  \right] = \left[ \begin{array}{ccc}
1 & 0 & 0 \\ 
1 & 1/2 & 1/4 \\
1 & 1 & 1
\end{array}  \right] \cdot \left[ \begin{array}{c}
c_1 \\ 
c_2 \\
c_3
\end{array}  \right] = A \cdot c.$$
Do đó
$$ c =  \left[ \begin{array}{c}

\end{array}  \right]= A^{-1} u = \left[ \begin{array}{ccc}

\end{array}  \right] \cdot \left[ \begin{array}{c}
u_1 \\ 
u_2 \\

\end{array}  \right] .$$
Từ đây ta thu được biểu diễn của trường nghiệm dưới dạng hàm bậc hai trong mỗi phần tử là:
\begin{eqnarray}
u(\xi) & = & c_1 + c_2 \xi + c_3 \xi^2 \\
& = & u_1 (1 - 3\xi + 2\xi^2) + u_2 (4\xi - 4\xi^2) + u_3(-\xi + 2\xi^2) \\
& = & u_1 \varphi_1(\xi) + u_2 \varphi_2(\xi) + u_3 \varphi_3(\xi),
\end{eqnarray}

trong đó: $\varphi_1(\xi) = 1 - 3\xi + 2\xi^2, \varphi_2(\xi) = 4\xi - 4\xi^2, \varphi_3(\xi) = -\xi + 2\xi^2$.

Bây giờ, ta lập luận tương tự như câu 1 ta thu được

\begin{equation}
\mathbf{u}(t + dt) = dt^2 (\mathbf{M}^T)^{-1}[\mathbf{f} - \mathbf{K}^T\mathbf{u}] + 2\mathbf{u} - \mathbf{u}(t-dt).
\end{equation}

trong đó, 
\begin{eqnarray}
&& u(t) \rightarrow u_i(t)\\

&& K \rightarrow K_{ij} = \int_{D_\xi} \mu \partial_\xi \varphi_i \partial_\xi \varphi_j \, d \xi \\

$$ M_{23}^i = M_{23}^i = \rho_i \int_0^1 (4\xi - 4\xi^2)(-\xi + 2\xi^2)\, d\xi  = \frac{1}{15} \rho_i.$$

Ta có ma trận độ cứng ở phần thử thứ i là  $K^i$ có dạng

\begin{equation}
K_{ij} = \mu
\begin{pmatrix}
\frac{7}{3} & -\frac{8}{3} &\frac{1}{3} &0 &0 &0 &\cdots &0 & 0 & 0 \\

0 &0 &\frac{1}{3} & -\frac{8}{3} & \frac{7}{3} + \frac{7}{3} & -\frac{8}{3} &\cdots &0 & 0 & 0\\
\vdots &\vdots &\vdots &\vdots &\vdots &\ddots &\vdots & \vdots & \vdots & \vdots\\
0 & 0 & 0 & 0 & 0 & 0 & \cdots  & \frac{7}{3} + \frac{7}{3} & -\frac{8}{3} &\frac{1}{3} \\
0 & 0 & 0 & 0 & 0 & 0 & \cdots & -\frac{8}{3} &\frac{16}{3} & -\frac{8}{3}\\
0 & 0 & 0 & 0 & 0 & 0 & \cdots & \frac{1}{3} & -\frac{8}{3} &\frac{7}{3}
\end{pmatrix}
\end{equation}

```python
# Import all necessary libraries, this is a configuration step for the exercise.
# Please run it before the simulation code!
import numpy as np
import matplotlib
# Show Plot in The Notebook
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python

# Initialization of setup
# ---------------------------------------------------------------
# Basic parameters
nx1 = 116
nx2 = 100 
nx3 = 230
nx    = nx1+nx2+nx3    # Number of collocation points  
xmax  = 10200.  # Physical dimension [m]
vs1    = 6000   # vận tốc miền thứ nhất [m/s] 
vs2 = 1500 # vận tốc miền thứ 2
vs3 = 3000  # vận tốc miền thứ 3
vs= np.arange(0,nx) # vecto vận tốc
for i in range(0,nx1):
    vs[i] = vs1
for i in range(nx1,nx1+nx2):
    vs[i] = vs2
for i in range(nx1+nx2,nx):
    vs[i] = vs3

ro0   = 2500    # Density [kg/m^3]
nt    = 18000   # Number of time steps
isx   = 223     # Source location [m] 
eps   = 0.5     # Stability limit
iplot = 20     # Snapshot frequency

#  dx = xmax/(nx-1)          
dx1 = 40     # calculate space increment
dx2 = 10
dx3 = 20
dx= np.arange(0,nx)
for i in range(0,nx1):
    dx[i] = dx1
for i in range(nx1,nx1+nx2):
    dx[i] = dx2
for i in range(nx1+nx2,nx):
    dx[i] = dx3
    
# initialize space coordinates
# x  = np.arange(0, nx)*dx
x  = np.arange(0, nx)
for i in range(0, nx1):
    x[i] = x[i]*dx[i]
for i in range(nx1, nx1+nx2):
    x[i] = x[nx1-1] +(i-nx1+1)*dx[i] 
for i in range(nx1+nx2, nx):
    x[i] = x[nx1+nx2-1]+ (i-nx1-nx2+1)*dx[i]
x  = np.transpose(x)

h = np.diff(x)  # Element sizes [m]

# parameters   
ro = x*0 + ro0

mu =  ro*vs**2
# time step from stabiity criterion
dt = 0.5*eps*np.min(dx[i])/np.max(np.sqrt(mu[i]/ro[i]))

# initialize time axis
t   = np.arange(1, nt+1)*dt 

# ---------------------------------------------------------------
# Initialize fields
# ---------------------------------------------------------------
u    = np.zeros(nx)
uold = np.zeros(nx)
unew = np.zeros(nx)

p    = np.zeros(nx)
pold = np.zeros(nx)
pnew = np.zeros(nx)
```

```python
# Mass matrix M_ij
# ---------------------------------------------------------------
M = np.zeros((nx,nx), dtype=float)
for i in range(1, nx):
    for j in range (1, nx):
        if j==i:
            M[2*int(i/2)-1,2*int(j/2)-1]= ro[i]*8/15
            M[2*int(i/2),2*int(j/2)]= ro[i]*4/15
        elif j==i+1:
            M[i,j] = ro[i]*1/15
        elif j==i-1:
            M[i,j] = ro[i]*1/15
        elif j==i+2:
            M[2*int(i/2)-1,2*int(j/2)-1 ]=0
            M[2*int(i/2),2*int(j/2)]=-ro[i]*1/30
        elif j==i-2:
            M[2*int(i/2)-1,2*int(j/2)-1 ]=0
            M[2*int(i/2),2*int(j/2)]= - ro[i]*1/30
        else:
            M[i,j] = 0
# Corner elements
M[0,0] = ro[i]*2/15
M[0,1] = ro[i]*1/15
M[0,2] = -ro[i]*1/30
M[1,0] = ro[i]*1/15
M[2,0] = -ro[i]*1/30
M[nx-1,nx-1] = ro[i]*2/15
# Invert M
Minv = np.linalg.inv(M)
Minv = np.linalg.inv(M)
```

```python
K = np.zeros((nx,nx), dtype=float)
for i in range(1, nx):
    for j in range (1, nx):
        if j==i:
            K[2*int(i/2)-1,2*int(j/2)-1]= mu[i]*16/3
            K[2*int(i/2),2*int(j/2)]= mu[i]*14/3
        elif j==i+1:
            K[i,j] = -mu[i]*8/3
        elif j==i-1:
            K[i,j] = -mu[i]*8/3
        elif j==i+2:
            K[2*int(i/2)-1,2*int(j/2)-1 ]=0
            K[2*int(i/2),2*int(j/2)]=mu[i]*1/3
        elif j==i-2:
            K[2*int(i/2)-1,2*int(j/2)-1 ]=0
            K[2*int(i/2),2*int(j/2)]= mu[i]*1/3
        else:
            K[i,j] = 0
            
# Corner elements
K[0,0] = mu[i]*7/3
K[0,1] = -mu[i]*8/3
K[0,2] = mu[i]*1/3
K[1,0] = -mu[i]*8/3
K[2,0] = mu[i]*1/3
K[nx-1,nx-1] = mu[i]*7/3
```
